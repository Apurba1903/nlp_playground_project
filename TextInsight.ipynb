```python
import nlpcloud


class NLPApp:
    
    # Constructor
    def __init__(self):
        
        self.__database = {}
        self.__first_menu()
    
    
    # First Menu Method    
    def __first_menu(self):
        
        first_input = input ("""
        Hi! How would you like to proceed?
        
        1. Register
        2. Login
        3. Exit        
        
        """)
        
        if first_input == '1':
            self.__register()
            
        elif first_input == '2':
            self.__login()
            
        else:
            exit()
    
    
    # Second Menu Method
    def __second_menu(self):
        
        second_input = input ("""
        Hi! How would you like to proceed?
        
        1. NER
        2. Language Detection
        3. Sentiment Analysis
        4. Logout        
        
        """)
        
        if second_input == '1':
            self.__ner()
        elif second_input == '2':
            self.__language_detection()
        elif second_input == '3':
            self.__sentiment_analysis()
        else:
            exit()
    
    
    # Registration Method
    def __register(self):
        name = input('Enter Name: ')
        email = input('Enter Email: ')
        password = input('Enter Password: ')
        
        
        if email in self.__database:
            print('Email already exists!')
        else:
            self.__database[email] = [name,password]
            print('Registration Successful. Please Login Now.')
            print(self.__database)
            self.__first_menu()
    
    
    # Login Method
    def __login(self):
        
        email = input('Enter Email: ')
        password = input('Enter Password: ')
        
        if email in self.__database:
            if self.__database[email][1] == password:
                print('Login Successful.')
                self.__second_menu()
            else:
                print('Wrong Password. Try Again.')
                self.__login()
            
        else:
            print('This email is not registered.')
            self.__first_menu()


    # NER Method
    def __ner(self):
        para = input("Enter the Paragraph: ")
        search_term = input('What would you like to search? : ')
        
        client = nlpcloud.Client("finetuned-llama-3-70b", "3884971768ff5f61fa20614d0ba0976962f2ce50", gpu=True)
        response = client.entities(para,searched_entity = search_term)
        
        print(response)
        self.__second_menu()
        
        
    
    # Language Detection Method
    def __language_detection(self):
        para = input("Enter the Paragraph: ")
        
        client = nlpcloud.Client("python-langdetect", "3884971768ff5f61fa20614d0ba0976962f2ce50", gpu=True)
        response = client.lang_detection(para)
        
        
        print(response)
        self.__second_menu()
    
    
        
    # Sentiment Analysis Method
    def __sentiment_analysis(self):
        para = input("Enter the Paragraph: ")
        
        client = nlpcloud.Client("distilbert-base-uncased-emotion", "3884971768ff5f61fa20614d0ba0976962f2ce50", gpu=False)
        response = client.sentiment(para, target="NLP Cloud")
        
        
        L = []
        for i in response['scored_labels']:
            L.append(i['score'])
        
        index = sorted(list(enumerate(L)), key = lambda x:x[1], reverse=True)[0][0]
        
        print('\n',response['scored_labels'][index]['label'])
        self.__second_menu()
        
        
obj = NLPApp()
```
